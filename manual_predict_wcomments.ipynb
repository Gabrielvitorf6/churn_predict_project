**Realizando novas predições com o nosso modelo serializado**
-

Essa é a maneira mais manual de realizar novas predições, porém não é a mais segura, pois dependemos do código para salvar o modelo, importar o modelo, importar os dados etc.

Isso significa que uma série de regras e caminhos precisam estar perfeitamente alinhados para a minha predição funcionar.

```
import pandas as pd

model_series = pd.read_pickle("model.pkl") #Leio o modelo
model = model_series['Model'] #Separo o modelo e as features em duas variáveis separadas
features = model_series['Features']
```

```
df = pd.read_csv("data/abt_churn.csv") #Lendo os dados da abt
df.head()
```

```
amostra = df[df['dtRef'] == df['dtRef'].max()].sample(3) #Criando amostra com as três observações mais recentes
amostra = amostra.drop('flagChurn', axis=1)
amostra.head()
```

```
predict = model.predict_proba(amostra[features])[:,1]
predict
```

```
amostra['proba'] = predict #Colocando a predição numa coluna nova do dataframe
amostra.head()
```

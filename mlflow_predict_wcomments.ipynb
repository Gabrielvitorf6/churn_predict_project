**Importando modelo registrado no MLFLow**
-

Essa é a versão mais segura e robusta de usar os nossos modelos treinados, pois deixa a cargo do mlflow gerenciar o modelo e armazenar as informações e versões dele.

```python
import mlflow
import pandas as pd
```

```python
mlflow.set_tracking_uri('http://localhost:5000')

model = mlflow.sklearn.load_model('models:/model_churn/1') #Passo o caminho do modelo, junto da versão do modelo que queremos importar. Como criamos só a primeira versão do modelo, passamos somente o número 1 no caminho
```

```python
dir(model) #Procurando o
```

```python
features = model.feature_names_in_
features
```

```python
model
```

```python
df = pd.read_csv("data/abt_churn.csv") #Lendo os dados da abt
df.head()
```

```python
amostra = df[df['dtRef'] == df['dtRef'].max()].sample(3) #Criando amostra com as três observações mais recentes
amostra = amostra.drop('flagChurn', axis=1)
amostra.head()
```

```python
predict = model.predict_proba(amostra[features])[:,1]
predict
```

```python
amostra['proba'] = predict #Colocando a predição numa coluna nova do dataframe
amostra.head()
```

Ele facilita também para trocarmos o modelo que está em produção, já que podemos simplesmente mudar a versão do modelo.

Vamos trocar o modelo para o que tem melhor AUC

```python
model = mlflow.sklearn.load_model('models:/model_churn/2') #Mudamos a versão do modelo model_churn para 2

predict = model.predict_proba(amostra[features])[:,1]
amostra['proba_modelv2'] = predict
amostra
```

Para melhorarmos MAIS ainda, podemos criar o código para que ele sempre use o último modelo registrado no mlflow, tirando a necessidade de alterar o caminho sempre que trocarmos o modelo em produção

```python
models = mlflow.search_registered_models(filter_string="name= 'model_churn'") #Procurando as informações dos modelos registrados, e filtrando o nome (nesse caso model_churn)

models[0] #Modelo registrado no mlflow. Nele temos as informações do nosso modelo num objeto/instância
```

```python
models[0].latest_versions
```

```python
latest_version = max([i.version for i in models[0].latest_versions]) #Percorro dentro das versões do objeto ModelVersion por meio de um loop for, armazeno as versões dentro de i.version e uso o max para pegar a maior versão (Ou a última, já que as versões são criadas de maneira crescente).
latest_version
```

```python
model = mlflow.sklearn.load_model(f'models:/model_churn/{latest_version}') #formatamos a última versão do modelo dentro do caminho, para que o script sempre use a versão mais recente do modelo

predict = model.predict_proba(amostra[features])[:,1]
amostra['proba_latestmodel'] = predict
amostra
```
